#distribuição de Bernoulli:

```r
p <- 0.7 #70% de chances de sucesso em um experimento
amostras <- rbinom(n = 1000, size = 1, prob = p)
print(amostras)
```

```r
tabela_freq <- table(amostras)

barplot(tabela_freq,
        col = c("red", "blue"),
        names.arg = c("Fracasso (0)", "Sucesso (1)"),
        main = "Distribuição de Bernoulli",
        ylab = "Frequencias",
        xlab = "Resultado")
```

#Distribuição de Poisson

Util quando desejamos obter o numero de ocorrencias em um intervalo especifico

```r
lambda <- 5 #Media de 5 eventos por intervalo
qtd_amostras <- 1000
amostras <- rpois(n = qtd_amostras, lambda = lambda )

print(amostras)

tabela_freq <- table(amostras)
barplot (tabela_freq,
          col="pink",
          main = "Distribuição de Poisson (lambda = 5)",
          ylab = "Frequência",
          xlab = "Npumero de ocorrências")
```
